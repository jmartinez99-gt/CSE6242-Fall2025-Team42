```python
pop = pd.read_pickle('../data/processed/Census.pkl')
pop = pop.rename(columns={'State': 'State Code', 'Population': 'State Population'})
pop
```

```python
merged = pd.merge(merged, pop, on=['Year', 'State Code'], how='inner')
merged
```

```python
merged['Total Enrollment'] = merged['Number of Grad Students'] + merged['Total Undergraduates']
merged = merged.rename(columns={'Number of Grad Students': 'Total Grad Students'})
merged[merged['Sanction Name']=='NCAA']
```

```python
merged.drop(columns=['INSTNM'])
front_cols = ['Year', 'UNITID', 'Institution Name', 'City', 'State Code', 'Region Name', 'School Type', 'Predominant Degree', 'Highest Degree', 'LONGITUDE', 'LATITUDE']
merged = merged[ front_cols + [col for col in merged.columns if col not in front_cols] ]
merged
```
