## Sport Participation

```python
import pickle
import pandas as pd
sp = pd.read_pickle("../data/processed/SportParticipation_AllYears.pkl")
sp
```

```python
sp_cols = list(sp.columns)
mens_part = [col for col in sp_cols if "Men's Team Part" in col]
mens_part
```

```python
womens_part = [col for col in sp_cols if "Women's Team Part" in col]
womens_part
```

```python
import re

coed_part = [col for col in sp_cols if re.search(r"Coed.*Part", col)]
coed_part
```

```python
total_participation_cols = [
    "Total Men's Team Participation",
    "Total Women's Team Participation",
    "Total Coed Team Men Participation",
    "Total Coed Team Women Participation",
    "Grand Total Men's Participation",
    "Grand Total Women's Participation",
    "Unduplicated Count Men's Participation",
    "Unduplicated Count Women's Participation"
]
```

```python
mens_revenue = [col for col in sp_cols if "Men's Team Rev" in col]
mens_revenue
```

```python
womens_revenue = [col for col in sp_cols if "Women's Team Rev" in col]
womens_revenue
```

```python
coed_revenue = [col for col in sp_cols if re.search(r"Coed.*Rev", col)]
coed_revenue
```

```python
mens_expenses = [col for col in sp_cols if "Men's Team Exp" in col]
mens_expenses
```

```python
womens_expenses = [col for col in sp_cols if "Women's Team Exp" in col]
womens_expenses
```

```python
coed_expenses = [col for col in sp_cols if re.search(r"Coed.*Exp", col)]
coed_expenses
```

```python
salary_cols = [
    c for c in sp_cols
    if "Average Annual Institutional Salary" in c
    or "Number of Head Coaches Included in Average" in c
    or "Number of Assistant Coaches Included in Average" in c
    or "Number of FTEs Included in Average" in c
]
salary_cols
```

```python
aid_cols = [
    c for c in sp_cols
    if "Athletic Student Aid" in c
]
aid_cols
```

```python
sp_dict = {
    "mens_part": mens_part,
    "womens_part": womens_part,
    "coed_part": coed_part,
    # "total_participation": total_participation_cols,
    "mens_revenue": mens_revenue,
    "womens_revenue": womens_revenue,
    "coed_revenue": coed_revenue,
    "mens_expenses": mens_expenses,
    "womens_expenses": womens_expenses,
    "coed_expenses": coed_expenses,
    # "salary": salary_cols,
    # "aid": aid_cols,
}
pickle.dump(sp_dict, open('../data/mappings/SportParticipation_columns.pkl', 'wb'))
```

```python
sp_combined_dict = {
    "all_part": mens_part + womens_part + coed_part,
    "all_revenue": mens_revenue + womens_revenue + coed_revenue,
    "all_expenses": mens_expenses + womens_expenses + coed_expenses
}
pickle.dump(sp_combined_dict, open('../data/mappings/SportParticipation_combined_columns.pkl', 'wb'))
```
