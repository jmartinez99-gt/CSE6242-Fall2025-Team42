```python
import pandas as pd
import pickle

df = pd.read_pickle('../data/merged/full_df.pkl')
df
```

```python
a_cols = pickle.load(open('../data/mappings/Accolades_columns.pkl' ,'rb'))
c_cols = pickle.load(open('../data/mappings/Census_columns.pkl' ,'rb'))
cs_dict = pickle.load(open('../data/mappings/CollegeScorecard_columns.pkl' ,'rb'))
cs_cols = [col for cols in cs_dict.values() for col in cols]
sp_dict = pickle.load(open('../data/mappings/SportParticipation_columns.pkl' ,'rb'))
sp_cols = [col for cols in sp_dict.values() for col in cols]
```

```python
import pandas as pd

all_knn_rows = []
knn_dict = pickle.load(open('../data/analysis/knn.pkl', 'rb'))

for knn_model, pred_dict in knn_dict.items():
    knn_df = df.copy()[cs_dict['keys'] + ['LONGITUDE', 'LATITUDE']].dropna(subset=['LONGITUDE'])
    for col, preds in pred_dict.items():
        knn_df[col] = preds.astype(int)
    knn_df['knn'] = knn_model
    all_knn_rows.append(knn_df)

knn_df_final = pd.concat(all_knn_rows, ignore_index=True)
```

```python
knn_df_final.to_csv(f'../data/tableau/knn.csv', index=False)
```
