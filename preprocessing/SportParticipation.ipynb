# SportParticipation

```python
import pandas as pd

df = pd.read_csv('../data/raw/All_Data_Combined_2004_2005_2006_2007_2008_2009_2010_2011_2012_2013_2014_2015_2016_2017_2018_2019_2020_2021_2022_2023.csv')
df
```

## Cleaning up missing data

```python
df.loc[df['Institution Name'] == 'Simon Fraser University', 'State CD'] = 'BC'
df[df['Total Undergraduates'].isna()].sort_values(["Institution Name", "Survey Year"]) # not sure if these should be omitted
```

```python
(df.isna() | df.eq('')).sum()
```

## Standardizing Leagues and Divisions

```python
division_name_map = {
    "NJCAA Division I": "Division 1",
    "Other": "Other",

    "NCAA Division III (with football)": "Division 3",
    "NCAA Division III with football": "Division 3",
    "NCAA Division III (without football)": "Division 3",
    "NCAA Division III without football": "Division 3",

    "NAIA Division I": "Division 1",
    "NAIA Division II": "Division 2",
    "NAIA Division III": "Division 3",

    "NJCAA Division II": "Division 2",
    "NJCAA Division III": "Division 3",

    "NCAA Division II (with football)": "Division 2",
    "NCAA Division II with football": "Division 2",
    "NCAA Division II (without football)": "Division 2",
    "NCAA Division II without football": "Division 2",

    "NCAA Division I-A": "Division 1",
    "NCAA Division I-AA": "Division 1",
    "NCAA Division I-AAA": "Division 1",
    "NCAA Division I-FBS": "Division 1",
    "NCAA Division I-FCS": "Division 1",
    "NCAA Division I without football": "Division 1",

    "CCCAA": "Other",
    "USCAA": "Other",
    "NWAC": "Other",
    "Independent": "Other",

    "NCCAA Division I": "Division 1",
    "NCCAA Division II": "Division 2",
}

has_football_map = {
    "NJCAA Division I": None,
    "NJCAA Division II": None,
    "NJCAA Division III": None,

    "NAIA Division I": None,
    "NAIA Division II": None,
    "NAIA Division III": None,

    "NCAA Division III (with football)": True,
    "NCAA Division III with football": True,
    "NCAA Division III (without football)": False,
    "NCAA Division III without football": False,

    "NCAA Division II (with football)": True,
    "NCAA Division II with football": True,
    "NCAA Division II (without football)": False,
    "NCAA Division II without football": False,

    "NCAA Division I-A": True,
    "NCAA Division I-FBS": True,
    "NCAA Division I-AA": True,
    "NCAA Division I-FCS": True,
    "NCAA Division I-AAA": False,
    "NCAA Division I without football": False,

    "CCCAA": None,
    "USCAA": None,
    "NWAC": None,
    "Independent": None,
    "Other": None,

    "NCCAA Division I": None,
    "NCCAA Division II": None,
}

football_subdivision_map = {
    "NCAA Division I-A": "FBS",
    "NCAA Division I-FBS": "FBS",

    "NCAA Division I-AA": "FCS",
    "NCAA Division I-FCS": "FCS",
}

insert_idx = df.columns.get_loc("Sanction Name") + 1
df.insert(insert_idx,     "Division Name",        "")
df.insert(insert_idx + 1, "Has Football",         None)
df.insert(insert_idx + 2, "Football Subdivision", "")

df["Division Name"] = df["Classification Name"].map(division_name_map).fillna("")
df["Has Football"] = df["Classification Name"].map(has_football_map)
df["Football Subdivision"] = (
    df["Classification Name"]
    .map(football_subdivision_map)
    .fillna("")
)
df
```

## Delete and shift columns

```python
# df = df.drop(columns=['UNITID', 'OPE ID', 'Sanction Code'])
df = df.drop(columns=['Sanction Code'])

cols = df.columns.tolist()
move_cols = ["Classification Name", "Classification Other"]
for c in move_cols:
    cols.remove(c)
insert_at = cols.index("Football Subdivision") + 1
new_cols = cols[:insert_at] + move_cols + cols[insert_at:]
df = df[new_cols]

df = df.rename(columns={'Survey Year': 'Year', "State CD": "State Code"})
df
```
