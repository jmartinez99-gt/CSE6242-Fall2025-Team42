```python
import pandas as pd

vi = pd.read_excel('../data/raw/POPTOTVIA647NWDB.xlsx', sheet_name='Annual')
vi['Year'] = list(range(2000, 2025))
vi['State'] = 'VI'
vi = vi.rename(columns={'POPTOTVIA647NWDB': 'Population'})
vi = vi.drop(columns=['observation_date'])
cols = [c for c in vi.columns if c != 'Population'] + ['Population']
vi = vi[cols]
vi = vi[(vi['Year'] > 2003) & (vi['Year'] < 2024)]
vi
```

```python
state_to_code = {
    "Alabama": "AL",
    "Alaska": "AK",
    "Arizona": "AZ",
    "Arkansas": "AR",
    "California": "CA",
    "Colorado": "CO",
    "Connecticut": "CT",
    "Delaware": "DE",
    "Florida": "FL",
    "Georgia": "GA",
    "Hawaii": "HI",
    "Idaho": "ID",
    "Illinois": "IL",
    "Indiana": "IN",
    "Iowa": "IA",
    "Kansas": "KS",
    "Kentucky": "KY",
    "Louisiana": "LA",
    "Maine": "ME",
    "Maryland": "MD",
    "Massachusetts": "MA",
    "Michigan": "MI",
    "Minnesota": "MN",
    "Mississippi": "MS",
    "Missouri": "MO",
    "Montana": "MT",
    "Nebraska": "NE",
    "Nevada": "NV",
    "New Hampshire": "NH",
    "New Jersey": "NJ",
    "New Mexico": "NM",
    "New York": "NY",
    "North Carolina": "NC",
    "North Dakota": "ND",
    "Ohio": "OH",
    "Oklahoma": "OK",
    "Oregon": "OR",
    "Pennsylvania": "PA",
    "Rhode Island": "RI",
    "South Carolina": "SC",
    "South Dakota": "SD",
    "Tennessee": "TN",
    "Texas": "TX",
    "Utah": "UT",
    "Vermont": "VT",
    "Virginia": "VA",
    "Washington": "WA",
    "West Virginia": "WV",
    "Wisconsin": "WI",
    "Wyoming": "WY",
    "Virgin Islands": "VI",
    "District of Columbia": "DC",
    "Puerto Rico": "PR"
}

us_pop = []
```

```python
df09 = pd.read_excel('../data/raw/st-est00int-01.xls', header=3)
df09 = (
    df09.rename(columns={'Unnamed: 0': 'State'})
    .drop(columns=['Unnamed: 1', 'Unnamed: 12', 'Unnamed: 13'])
)
us_pop += df09.loc[df09['State'] == 'United States'].values.tolist()[0][5:]
remove_states = ["United States", "Northeast", "Midwest", "South", "West"]
df09 = df09[~df09['State'].isin(remove_states)]
df09['State'] = df09['State'].str.lstrip('.')
df09['State'] = df09['State'].map(state_to_code)
df09 = df09.dropna(subset=['State'])
df09 = df09.reset_index(drop=True)
df09
```

```python
df09_long = df09.melt(
    id_vars='State',
    value_vars=[2004,2005,2006,2007,2008,2009],
    var_name='Year',
    value_name='Population'
)
df09_long['Year'] = df09_long['Year'].astype(int)
df09_long
```

```python
df20 = pd.read_excel('../data/raw/nst-est2020.xlsx', header=3)
df20 = (
    df20.rename(columns={'Unnamed: 0': 'State'})
    .drop(columns=['Census', 'Estimates Base', 'April 1', 'July 1'])
)
us_pop += df20.loc[df20['State'] == 'United States'].values.tolist()[0][1:]
remove_states = ["United States", "Northeast", "Midwest", "South", "West"]
df20 = df20[~df20['State'].isin(remove_states)]
df20['State'] = df20['State'].str.lstrip('.')
df20['State'] = df20['State'].map(state_to_code)
df20 = df20.dropna(subset=['State'])
df20 = df20.reset_index(drop=True)
df20
```

```python
df20_long = df20.melt(
    id_vars='State',
    value_vars=list(range(2010, 2020)),
    var_name='Year',
    value_name='Population'
)
df20_long['Year'] = df20_long['Year'].astype(int)
df20_long
```

```python
df24 = pd.read_excel('../data/raw/NST-EST2024-POP.xlsx', header=3)
df24 = (
    df24.rename(columns={'Unnamed: 0': 'State'})
    .drop(columns=['Unnamed: 1'])
)
us_pop += df24.loc[df24['State'] == 'United States'].values.tolist()[0][1:-1]
remove_states = ["United States", "Northeast", "Midwest", "South", "West"]
df24 = df24[~df24['State'].isin(remove_states)]
df24['State'] = df24['State'].str.lstrip('.')
df24['State'] = df24['State'].map(state_to_code)
df24 = df24.dropna(subset=['State'])
df24 = df24.reset_index(drop=True)
df24
```

```python
df24_long = df24.melt(
    id_vars='State',
    value_vars=list(range(2020, 2024)),
    var_name='Year',
    value_name='Population'
)
df24_long['Year'] = df24_long['Year'].astype(int)
df24_long
```

```python
all_years = pd.concat([df09_long, df20_long, df24_long], ignore_index=True)
all_years
```

```python
dfus = pd.DataFrame({'Year': list(range(2004, 2024)), 'US Population': us_pop})
dfus
```

```python
merged_df = pd.merge(all_years, dfus, on='Year', how='inner')
merged_df
```

```python
merged_df.to_pickle('../data/processed/state_populations.pkl')
```
