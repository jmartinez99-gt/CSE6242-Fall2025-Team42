```python
import requests
import zipfile
import io
import pandas as pd

# Example: one-file download; you’d need to replace with the direct Zip URL for the year you want
EADA_ZIP_URL = "https://public.tableau.com/89c6ed83-c06c-4cf9-ba1f-493f4cbde1e5"


resp = requests.get(EADA_ZIP_URL)
resp.raise_for_status()

with zipfile.ZipFile(io.BytesIO(resp.content)) as z:
    print("Files in EADA zip:", z.namelist())
    # Pick the CSV/Excel inside (inspect z.namelist())
    csv_name = [name for name in z.namelist() if name.lower().endswith(".csv")][0]
    with z.open(csv_name) as f:
        eada_df = pd.read_csv(f)
        print(eada_df.head())
```

```python
from tableauscraper import TableauScraper as TS

url = "https://public.tableau.com/89c6ed83-c06c-4cf9-ba1f-493f4cbde1e5"


ts = TS()
ts.loads(url)

workbook = ts.getWorkbook()

print("=== Worksheets ===")
for ws in workbook.worksheets:
    print(ws.name)

print("\n=== Preview of first worksheet ===")
first = workbook.worksheets[0]
print(first.data.head())

print("\n=== Filters in this worksheet ===")
for f in first.getFilters():
    print(f)

#<img draggable=​"false" style=​"left:​ 0px;​ top:​ 0px;​ max-width:​ 1360px;​ max-height:​ 409px;​" data-datasrc=​"https:​/​/​public.tableau.com/​vizql/​tilecache/​9EAB8B0CF1ED462184A85B9DEBAF55DF-0:​0/​9100/​a6fe0124c4fc72a3cff77d5acafcbadc744d2bb23a32c89bfbb0ba63cb04d1eb/​views.1961142049571216793_17075098080766875869.viz.0.0.png?​=1763420270994Z3" src=​"blob:​https:​/​/​public.tableau.com/​89c6ed83-c06c-4cf9-ba1f-493f4cbde1e5">​Event
```
