Cluster using sports accolades

```python
import pandas as pd

df = pd.read_pickle('../data/merged/full_df.pkl')
df
```

```python
import pickle
a_cols = pickle.load(open('../data/mappings/Accolades_columns.pkl' ,'rb'))
c_cols = pickle.load(open('../data/mappings/Census_columns.pkl' ,'rb'))
cs_dict = pickle.load(open('../data/mappings/CollegeScorecard_columns.pkl' ,'rb'))
sp_dict = pickle.load(open('../data/mappings/SportParticipation_columns.pkl' ,'rb'))
```

```python
expense_cols = sp_dict['mens_expenses'] + sp_dict['womens_expenses']
expense_cols
```

```python
from sklearn.cluster import KMeans

X = df[expense_cols].fillna(0).values > 0
```

```python
preds = {}
for k in list(range(3, 10, 2)):
    kmeans = KMeans(
        n_clusters=k,
        random_state=6740,
        n_init=10
    )
    preds[f'cluster (k={k})'] = kmeans.fit_predict(X)
preds
```

```python
pickle.dump(preds, open('../data/analysis/clustering.pkl', 'wb'))
```
